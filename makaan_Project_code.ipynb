# Data Cleaning

### Type conversion

```python
df['Price per sq/ft (₹)']=df['Price per sq/ft (₹)'].astype(int)
df['Area(sq/ft)']=df['Area(sq/ft)'].astype(int)
```

### Creating a Price column and its Type conversion

```python
df['Price']=df['Price per sq/ft (₹)']*df['Area(sq/ft)']
df['Price']=df['Price'].astype(float)
```

### Replacing

```python
df['Type']=df['Type'].replace('Plot','Residential')
#df['Locality']=df['Locality'].replace('Pragathi Nagar Kukatpally','Kukatpally')
df['Type']=df['Type'].replace('House','Independent House')
```

### Reindexing 

```python
new=['BHK','Type','Price','L/Cr','Area(sq/ft)','Price per sq/ft (₹)','Construction Status','Locality','City']
df=df[new]
df.reindex(columns=new)
df
```

### Deleting a Column 

```python
df.drop('L/Cr',axis=1,inplace=True)
```

### Renaming a Column

```python
df.rename(columns={'Price':'Price(in Crores)'},inplace=True)
```

### Changing Values in a Column

```python
df['Price(in Crores)']=round(df['Price(in Crores)']/10000000,3)
```
